```python
import docx
```

Python-docx allows you to create new documents as well as make changes to existing ones.

## Opening a document

```python
from docx import Document

document = Document()
document.save('test.docx')
```

This creates a new document from the built-in default template and saves it unchanged to a file named ‘test.docx’. The so-called “default template” is actually just a Word file having no content, stored with the installed python-docx package.

## REALLY opening a document

If you want more control over the final document, or if you want to change an existing document, you need to open one with a filename:
1. document = Document('existing-document-file.docx')
2. document.save('new-file-name.docx')

Q. Write a Python program to extract data from the MS Word documents using python( please provide the path of the document file according to the specific location )

```python
doc = docx.Document(r"C:\Users\bhaka\Documents\NLP AS-2.docx")
```

```python
# creating an empty list
fullText=[]
```

```python
for para in doc.paragraphs:
 fullText.append(para.text)
print('\n'.join(fullText))

# We are iterating through paragraphs in a document and appending them to a list called fullText, and then printing the entire fullText list.
```
