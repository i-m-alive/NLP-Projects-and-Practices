```python
import pandas as pd

# Load CSV file
df = pd.read_csv('/content/drive/MyDrive/Project/project dataset/insparation.csv')
```

```python
Quotes = df['Quote']
```

```python
Quotes
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

The Tokenizer class in tensorflow.keras.preprocessing.text module is used to vectorize a text corpus into a list of integers. Each integer represents the index of a word in a vocabulary dictionary. It provides methods to tokenize a text corpus, create a word index, and convert text data into sequences of integers.

```python
tokenizer= Tokenizer()
```

```python
tokenizer.fit_on_texts(Quotes)
```

```python
#to view the words with their token names
tokenizer.word_index
```

```python
len(tokenizer.word_index)# total no. of unique words
```

```python
#sentence with their word sequences represented as numbers or say vectors
for sentence in Quotes:
    print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_sequences = []
for sentence in Quotes:
    tokenized_sentence= (tokenizer.texts_to_sequences([sentence])[0])

    for i in range(1,len(tokenized_sentence)):
       input_sequences.append(tokenized_sentence[:i+1])
#perfroming  slicing(tokenized_sentence[:i+1]) and appending to a new list
```


this will replace all the words with their respective index no. or say token no in each and every line. for trainng the model for text generation we need to have a dataset which will give input as the sequence of the sentence one word at a time and the output as the next words after each entry


```python
input_sequences
```

```python
#we will find the length of the longest list so that we can apply padding
max_len = max(len(x) for x in input_sequences)
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen = max_len,padding="pre")
```

For perfroming paddding there is already a function in keras so we will import that function

```python
padded_input_sequences
```

```python
x=padded_input_sequences[:,:-1] #using slice operation
```

```python
y=padded_input_sequences[:,-1]
```

```python
x,y
```

```python
 x.shape
```

```python
 y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=3764)
#used to perform One hot encoding
```

```python
y
```

```python
y.shape
```

```python
#importing necessery libraries
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
```

```python

# Define the model
model = Sequential()

# Add an Embedding layer
model.add(Embedding(input_dim=3764, output_dim=200, input_length=114))

# Add LSTM layers
model.add(LSTM(units=150, return_sequences=True))  # First LSTM layer returning sequences
model.add(LSTM(units=150))  # Second LSTM layer not returning sequences

# Add Dense layer
model.add(Dense(3764, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=100,verbose=1)
```

```python
text='about'

#tokenize
token_text=tokenizer.texts_to_sequences([text])[0]
#padding
padded_input=pad_sequences([token_text],maxlen=114,padding='pre')
print(padded_input)
```

```python
#predict
predict =model.predict(padded_input)
import numpy as np
p=np.argmax(predict)
```

```python
for word, index in tokenizer.word_index.items():
  if index == p:
    print(word)
```

```python
# for predicting 5 words
text="about"
for i in range(5):
  #tokenize
  token_text=tokenizer.texts_to_sequences([text])[0]
  #padding
  padded_input=pad_sequences([token_text],maxlen=114,padding='pre')
  #predict
  predict =np.argmax(model.predict(padded_input))
  for word, index in tokenizer.word_index.items():
    if index == predict:
      text = text + " "+ word
      print(text)
```
